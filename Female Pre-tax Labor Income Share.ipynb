```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ineq_data = pd.read_csv("WID_Data.csv", sep=";", skiprows=1)
```

```python
ineq_data.columns = ['Percentile', 'Year', 'female income share', 'CN_income']
```

```python
ineq_data.plot(x='Year', y=['female income share'])
plt.title("female pre-tax labor income share", fontsize= 15)
plt.xlabel("Year",fontsize = 12)
plt.ylabel("value",fontsize = 12)
plt.show()
```

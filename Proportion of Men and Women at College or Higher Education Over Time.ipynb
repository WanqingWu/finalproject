```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load the dataset
data = pd.read_csv("men_and_women_total.csv")
```

```python
plt.title("Proportion of men and women at college or higher education over time")
plt.plot(data['Year'], data['Male'], label='Male', color='black')
plt.plot(data['Year'], data['Female'], label='Female', color='blue')
plt.legend()
plt.scatter(data['Year'], data['Male'], color='black')
plt.scatter(data['Year'], data['Female'], color='blue')
plt.xticks(np.arange(1999, 2020, step=1))
plt.xticks(rotation=45, fontsize=10)
plt.show
```
